```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("data/housing.csv")
df.head()
```

```python
df.shape
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.histplot(df["median_house_value"], bins=50, kde=True)
plt.show()
```

```python
sns.scatterplot(x=df["median_income"], y=df["median_house_value"], alpha=0.3)
plt.show()
```

```python
df.corr(numeric_only=True)["median_house_value"].sort_values(ascending=False)
```

```python
median = df["median_house_value"].median()

df["price_category"] = np.where(
    df["median_house_value"] > median,
    "expensive",
    "cheap"
)

df["price_category"].value_counts()
```
